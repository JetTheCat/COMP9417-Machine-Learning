```python
import pandas as pd
from sklearn import preprocessing
import argparse
import lib.nnb
import matplotlib.pyplot as plt
import seaborn as sns

# Gets training data, randomises and seperates into X,Y values.
train = pd.read_csv("data/train.csv").sample(frac=1)
trainX = train.iloc[:,1:-1]     # remove ID and Label 
trainY = train.iloc[:,-1]

# Gets test data and seperates into X,Y values.
test = pd.read_csv("data/train.csv")
testX = test.iloc[:,1:-1]       # remove ID and Label 
testY = test.iloc[:,-1]


# Normalise training data from 0<=X<=1
x = trainX.values
min_max_scalar = preprocessing.MinMaxScaler()
train_scaled = min_max_scalar.fit_transform(x)
test_scaled = min_max_scalar.fit_transform(x)
scaled_trainX = pd.DataFrame(train_scaled)
scaled_testX = pd.DataFrame(test_scaled)



# Function to save the predictions to a Kaggle submission file
def save_for_submit(ids, targets, filename):
    df = pd.DataFrame()
    df['id'] = ids
    for i in range(1, 10):
        col = 'Class_' + str(i)
        df[col] = list(map(lambda x: 1 if x == col else 0, targets))
    df.to_csv(filename, index=False)
```

# Pearson Selection for Feature Selection

```python
# Converting target/output column values to numerical representation
target = train.iloc[:,-1]
label_enc = preprocessing.LabelEncoder()
label_enc.fit(target)
target = label_enc.transform(target)
target = pd.DataFrame(target)
new_train = scaled_trainX.copy(deep=True)
new_train['target'] = target

# Showing Pearson Selection for Feature Selection 
plt.figure(figsize=(95,95))
cor = new_train.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
#Correlation with output variable
cor_target = abs(cor["target"])
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.3]

#Remove matching corelated feature
relevant_features = relevant_features[:-1]

# Build new dataframe with relevant feature
new_trainX = pd.DataFrame()
new_testX = pd.DataFrame()
feat_index = list(relevant_features.index)

new_trainX = scaled_trainX[feat_index]
    
new_testX = new_trainX.copy()
```

## Implementing Classifier with Relevant Features

```python
relevant_features
```

```python
# Multinomial Naive Bayes Implementation

from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

pred = pd.read_csv("data/test.csv").sample(frac=1)
pred = pred.iloc[:,1:] # remove ID and Label 
pred = min_max_scalar.fit_transform(pred)
pred = pd.DataFrame(pred)

clf = MultinomialNB()
clf.fit(new_trainX, trainY)
print("Accuracy Score: {}".format(clf.score(new_testX, testY)))

df_test = pd.read_csv('data/test.csv')
tmp = df_test.drop(columns=['id'])

# Normalise test data from 0<=X<=1
x = tmp
min_max_scalar = preprocessing.MinMaxScaler()
test_scaled = min_max_scalar.fit_transform(x)
X_test = pd.DataFrame(test_scaled)

X_test = X_test[feat_index]
mnb_pred = clf.predict(X_test)
#save_for_submit(df_test.id, mnb_pred, 'multinb_pearson_output.csv')
```

```python
from sklearn.metrics import confusion_matrix

y_pred = clf.predict(new_testX)
cm = confusion_matrix(testY, y_pred)

plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True, fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

```python
# Logistic Regression Implementation
from sklearn.linear_model import LogisticRegression

le_train = preprocessing.LabelEncoder()
le_test = preprocessing.LabelEncoder()
le_train.fit(trainY)
trainY = le_train.transform(trainY)
le_test.fit(testY)
testY = le_test.transform(testY)


clf = LogisticRegression(random_state=0, max_iter=6000)
clf.fit(new_trainX, trainY)
print("Accuracy Score: {}".format(clf.score(new_testX, testY)))
```

```python
y_pred = clf.predict(new_testX)
cm = confusion_matrix(testY, y_pred)

plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True, fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

## Pearson Selection Conclusion

We can observe that we are unable to achieve an accuracy score > 30% with correlated features that scored > 0.3. We are also unable to find any correlated features that scored > 0.5 and by definition of what a correlation coefficient represent below we can conclude this selection method is not viable.

The correlation coefficient has values between -1 to 1
* A value closer to 0 implies weaker correlation (exact 0 implying no correlation)
* A value closer to 1 implies stronger positive correlation
* A value closer to -1 implies stronger negative correlation

# Backward Elimination (Wrapper Method for Feature Selection)

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

#Adding constant column of ones, mandatory for sm.OLS model
X_1 = sm.add_constant(scaled_trainX)
#Fitting sm.OLS model
model = sm.OLS(trainY,X_1).fit()
model.pvalues

X_1.head()
```

```python
cols = list(scaled_trainX.columns)
pmax = 1
while (len(cols)>0):
    p= []
    X_1 = scaled_trainX[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(trainY,X_1).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```

```python
be_trainX = scaled_trainX.copy(deep=True)
be_trainX = be_trainX[selected_features_BE]
be_testX = be_trainX.copy()
```

```python
clf = LogisticRegression(random_state=0, max_iter=6000)
clf.fit(be_trainX, trainY)
print("Accuracy Score: {}".format(clf.score(be_testX, testY)))
```

```python
y_pred = clf.predict(be_testX)
cm = confusion_matrix(testY, y_pred)

plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True, fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

## Backward Elimination Conclusion

We can observe that we are unable to achieve an accuracy score > 20% with the given selected features, therefore this selection method is also not viable to help improve the model prediction.
